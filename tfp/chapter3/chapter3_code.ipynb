```python
# imports
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_probability as tfp
tfd = tfp.distributions
import numpy as np
import arviz as az
```

```python
Pr_Positive_Vampire = tf.constant(0.95)
Pr_Positive_Mortal = tf.constant(0.01)
Pr_Vampire = tf.constant(0.001)
Pr_Positive = Pr_Positive_Vampire * Pr_Vampire + Pr_Positive_Mortal * ( 1 - Pr_Vampire )
Pr_Vampire_Positive = Pr_Positive_Vampire * Pr_Vampire / Pr_Positive
print(Pr_Vampire_Positive.numpy())
```

R code 3.2 equivalent

```python
@tf.function
def grid_approx(grid_size, num_draws, num_successes):
    # define grid
    #grid_size = tf.constant(size_of_grid)
    p_grid = tf.linspace(start=0., stop=1., num=grid_size)
    p_grid = tf.expand_dims(p_grid, 1)
    # define prior
    prior = tf.ones([grid_size, 1])
    # compute likelihood at each value in grid
    likelihood = tfd.Binomial(num_draws, probs=p_grid).prob(num_successes)
    # compute product of likelihood and prior
    unstd_posterior = likelihood * prior
    # standardize the posterior, so it sums to 1
    posterior = unstd_posterior / tf.math.reduce_sum(unstd_posterior)
    return p_grid, posterior
# generate posterior
grid_size = tf.constant(1000)
num_draws = 9
num_successes = 6
(p_grid, posterior) = grid_approx(grid_size, num_draws, num_successes)
print(tf.size(posterior))
```

R code 3.3 equivalent

```python
@tf.function
def generate_posterior_samples(num_samples, posterior, grid_size):
    samples = tfd.Categorical(probs=tf.reshape(posterior, [grid_size])).sample(num_samples)
    posterior_samples = samples / (grid_size - 1)
    #samples = tfd.Multinomial(total_count=1, probs=tf.reshape(posterior, [grid_size])).sample(num_samples)
    #indices = tf.where(samples)
    #posterior_samples = indices[:,1] / (grid_size - 1)
    return posterior_samples

num_samples = tf.constant(10000)
posterior_samples = generate_posterior_samples(num_samples, posterior, grid_size)
```

```python
# need to regenerate posterior and posterior samples (in the meantime we did a 3 draws from 3 trials example)
# this is the bottom subfigure in Fig 3.6 (for each sample--a value for the probability of water--in 
# posterior_samples, it generates a random binomial prediction for the number of water samples from 9 tosses)

grid_size = 1000
num_tosses = 9
num_successes = 6
num_samples = 1e4

(_, posterior) = grid_approx(grid_size, num_tosses, num_successes)
posterior_samples = generate_posterior_samples(num_samples, posterior, grid_size)

w = tfd.Binomial(9, probs=posterior_samples).sample(1)
_, ax = plt.subplots()
ax.hist(w[0,:].numpy(), bins=10)
ax.set_xlabel("number of water samples")
ax.set_ylabel("posterior predictive distribution");
```
